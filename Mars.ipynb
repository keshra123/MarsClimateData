```python


def TheParser(path):
  temp = ''
  pre = ''
  c = 0
  with open(path) as a:
    line = a.readline()
    while line:
      if line.startswith("### Retrieved on:"):
        date = line.split(':', 1)[1].split('T')[0]
      if line.startswith('#'):
        line = a.readline()
      if line.startswith('-----------------------'):
        print('here')
        line = a.readline()
      if line.startswith('----'):
        line = line.replace("----",date)
        print("---here")
        c +=1
      if not line.startswith('#'):
          if c == 1:
            temp = temp+line
          if c ==2:
            pre = pre + line
          line = a.readline()
     
    
  temp = temp.replace('    ',',').replace('||','').replace('   ',',').replace(' ','')
  pre = pre.replace('    ', ',').replace('||', '').replace('   ',',').replace(' ','')
  import pandas as pd
  import io
  buffer = io.StringIO(temp)
  r=pd.read_csv(buffer)
  df = pd.DataFrame(r)
  
  df =df.T
  df = df.reset_index()
  df.to_csv(path.replace('.txt','.csv').replace('temperature','month'))
```
